```python
# using JSON file for my api_key

# Need to import json to do this
# Need to requests 
# Need to pandas for dataframe
import json
import requests
import pandas as pd

# Define a function for openning the json
def get_keys(path):
    with open(path) as f:
        return json.load(f)

# Using the function to open and load all keys in that file 
api_keys = get_keys("/Users/ezgi/.secret_the_movie/movie_api.json")

# Setting the first (and only) value as a variable
ds_key_1 = list(api_keys.values())[0]


```

```python
df = pd.DataFrame.from_dict(ds_r.json()['results'])
print(len(df)) #Print how many rows
print(df.columns) #Print column names
df
```

```python
highest_revenue = ds_r.json() # store parsed json response

# uncomment the next line to get a peek at the highest_revenue json structure
# highest_revenue

highest_revenue_films = highest_revenue['results']
```

```python
columns = ['id','film', 'revenue','genres','vote_count','vote_average','release_date','original_language',
           'orijinal_title','popularity']


# create dataframe with film and revenue columns
df_1= pd.DataFrame(columns=columns)
```

```python
for film in highest_revenue_films:
    # print(film['title'])
    film_revenue = requests.get('https://api.themoviedb.org/3/movie/'+ str(film['id']) +'?api_key='+ ds_key_1+'&language=en-US')
    film_revenue = film_revenue.json()
    #print(locale.currency(film_revenue['revenue'], grouping=True ))
    df_1.loc[len(df_1)]=[film['id'],film['title'],film_revenue['revenue'],film['genre_ids'],film['vote_count'],
                         film['vote_average'],film['release_date'],film['original_language'],film['original_title'],
                          film['popularity']]


# create dataframe with film and revenue columns
df_1
```
