# Question 3: What is the best month for releasing the movie?

In this question, we will figure out the best time for new movie according to best revenue.

1. Clean 'Worldwide' column
2. Clean 'release_date' columns
3. Take the month values
4. See the general common release months
5. See the best months for highest revenue

```python
import pandas as pd #import for dataframe
```

```python
df = pd.read_csv('last_all_data.csv') #read our last data
```

```python
df #to see the dataframe
```

```python
df.dropna(subset=['release_date'],inplace=True) #drop null values for date
```

```python
df.dropna(subset=['Worldwide'],inplace=True) #drop null values for worldwide
```

# Clean the Worldwide Column

```python
df['Worldwide']=df['Worldwide'].str.replace('$','') #Clean $ values
```

```python
df['Worldwide']=df['Worldwide'].str.replace(',','') #Clean (,) values
```

```python
df['Worldwide']=df['Worldwide'].astype(int) #change the type to integer
```

```python
df.info() # to check the columns types
```

# Changing the Date Column to Datetime and Take Month in a New Column

```python
import datetime #for changing string to date-time
df['released_month'] = pd.to_datetime(df['release_date']) #create new datetime type column
```

```python
df['month'] = pd.DatetimeIndex(df['released_month']).month #take the month in a new column
```

```python
import calendar #to change the number of months to names of months
df['month'] = df['month'].apply(lambda x: calendar.month_abbr[x])
```

```python
df['month'].value_counts() #to see the most common
```

```python
import numpy as np     #import items for plotting

import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
%matplotlib inline
```

# Plot the Common Release Months 

```python
month_count  = df['month'].value_counts() #for taking x and y values easily

plt.figure(figsize=(20,8))    #set the figure size
sns.barplot(month_count.index, month_count.values, alpha=0.8) #plot with seaborn
plt.title('Released Months',fontsize=30) #set the title
plt.ylabel('Number of Movies', fontsize=20) #y label name
plt.xlabel('Months', fontsize=20) #x label name
plt.xticks(fontsize=25) # setting the font size of values of x-axis
plt.xticks(rotation=45) #rotate the x-axis labels
plt.yticks(fontsize=25) # setting the font size of values of y-axis
plt.show()
```

# Best Month for Release According to Revenue > 1000000000

```python
best_gross_df = df[df['Worldwide'] >= 1000000000] #put the condition to find higher than 1000000000$
```

```python
best_gross_df['month'].value_counts() #see the best months
```

```python
month_count1  = best_gross_df['month'].value_counts()

plt.figure(figsize=(20,8)) #set the figure size
sns.barplot(month_count1.index, month_count1.values, alpha=0.8) #plot with seaborn
plt.title('Released Months of Top Worldwide Gross',fontsize=30) #set the title
plt.ylabel('Number of Movies', fontsize=20) #y label name
plt.xlabel('Months', fontsize=20) #x label name
plt.xticks(fontsize=25) # setting the font size of values of x-axis
plt.xticks(rotation=45) #rotate the x-axis labels
plt.yticks(fontsize=25) # setting the font size of values of y-axis
plt.show()
```
