```python
import json
import requests
import pandas as pd

# Define a function for openning the json
def get_keys(path):
    with open(path) as f:
        return json.load(f)

# Using the function to open and load all keys in that file 
api_keys = get_keys("/Users/ezgi/.secret_the_movie/movie_api.json")

# Setting the first (and only) value as a variable
ds_key_1 = list(api_keys.values())[0]
```

```python
# url = f'https://api.themoviedb.org/3/discover/movie?api_key={ds_key_1}&primary_release_year=2019&sort_by=revenue.des

url = f'https://api.themoviedb.org/3/movie/top_rated?api_key={ds_key_1}&language=en-US&page=1'
                     
              
              
              
              # for yearly highest revenue for 2019
```

```python
ds_r = requests.get(url)
```

```python
df = pd.DataFrame.from_dict(ds_r.json()['results'])
print(len(df)) #Print how many rows
print(df.columns) #Print column names
df
```
